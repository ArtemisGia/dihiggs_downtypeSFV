```python
import os
import pandas as pd 
import numpy as np 
import csv
import matplotlib.pyplot as plt
import matplotlib
```

```python
BrHZZ = pd.read_csv('atlas_BR_H_to_ZZ_ks.csv')
BrHZZ['mH'] = BrHZZ['mH'].astype(float)
BrHZZ.head()
```

```python
cmsBrHZZ = pd.read_csv('cms_BR_H_to_ZZ_ks.csv')
cmsBrHZZ['mH'] = cmsBrHZZ['mH'].astype(float)
cmsBrHZZ.head()
```

```python
#Copy the {mh2,ks,xsec} results from MG5 Events directory to a dataframe labeled simres
ks = []
mh2 = []
xs = []
# Specify the directory you want to search
root_directory = '/gpfs/scratch/argiannakopo/H_to_ZZ_ks_cosbma0p1_xi0_atlas/Events/'

# Use os.walk to iterate through all subdirectories
for root, dirs, files in os.walk(root_directory):
    for subdir in dirs:
        # Print the full path of each subdirectory
        subdirectory_path = os.path.join(root, subdir)
        os.chdir(subdirectory_path)
    # with gzip.open('unweighted_events.lhe.gz', 'rb') as f_in:
    #     with open('unweighted_events.LHE', 'wb') as f_out:
    #         shutil.copyfileobj(f_in, f_out)
        start1 = 0
        start2 = 0
        with open(subdir+'_tag_1_banner.txt', 'r') as csvfile:

            rows = csv.reader(csvfile, skipinitialspace=True, delimiter=' ')
            for row in rows:
                if len(row) > 1:
                    if row[0] == "BLOCK":
                        start1 = 1
                    if start1 == 1:
                        if row[2] == "#" and row[3] == "ks":
                            ks.append(float(row[1]))
                        if row[0] == "35" and row[3] == "mh2":
                            mh2.append(float(row[1]))
                    if row[1] == "Integrated" and row[2] == "weight":
                        xs.append(float(row[5]))

simres = pd.DataFrame(zip(mh2,ks,xs), columns = ['mh2','ks','xs'])
simres.head()
```

```python
simres['mh2'] = simres['mh2'].round(4)
simres['ks'] = simres['ks'].round(4)
BrHZZ['mH'] = BrHZZ['mH'].round(4)
BrHZZ['ks'] = BrHZZ['ks'].round(4)
cmsBrHZZ['mH'] = cmsBrHZZ['mH'].round(4)
cmsBrHZZ['ks'] = cmsBrHZZ['ks'].round(4)
```

```python
# NWA: Multiply each row of the simulation results dataframe with the appropriate Br(A -> Zh)

NWA = pd.DataFrame(columns=['mh2', 'ks', 'ZZ', 'xs'])
for i in range(len(simres)):
    mh2 = simres.at[i, 'mh2']
    ks = simres.at[i, 'ks']
    
    matching_rows = BrHZZ[(BrHZZ['mH'] == mh2) & (BrHZZ['ks'] == ks)]
    
    if not matching_rows.empty:
        HZZ = matching_rows['ZZ'].values[0].astype(float)
        xsec = simres.at[i, 'xs']
        NWA.loc[i, 'xs'] = HZZ * xsec
        NWA.loc[i, 'mh2'] = mh2
        NWA.loc[i, 'ks'] = ks
        NWA.loc[i, 'ZZ'] = HZZ
        
for i in range(len(simres)):
    mh2 = simres.at[i, 'mh2']
    ks = simres.at[i, 'ks']
    
    matching_rows = cmsBrHZZ[(cmsBrHZZ['mH'] == mh2) & (cmsBrHZZ['ks'] == ks)]
    
    if not matching_rows.empty:
        HZZ = matching_rows['ZZ'].values[0].astype(float)
        xsec = simres.at[i, 'xs']
        NWA.loc[i, 'xs'] = HZZ * xsec
        NWA.loc[i, 'mh2'] = mh2
        NWA.loc[i, 'ks'] = ks
        NWA.loc[i, 'ZZ'] = HZZ
```

```python
NWA = NWA[NWA['xs'] != 0 ].reset_index(drop=True)
NWA.head()
```

```python
# Load the experimental bounds to compare with my results 

ATLAS = pd.read_csv('/gpfs/home/argiannakopo/SFV/Exp_data/X-ZZ-ATLAS-2009p14791.csv')
ATLAS.head()
```

```python
# Load the experimental bounds to compare with my results 

CMS = pd.read_csv('/gpfs/home/argiannakopo/SFV/Exp_data/X-ZZ-CMS-1804p01939.csv')
CMS.head()
```

```python
### Interpolate based on the values of the narrow width approximation to get final bound

data_rows = []

for m in ATLAS['mH']:
    chunk = pd.DataFrame(columns=['mh2', 'ks', 'xs'])
    for i in range(len(NWA)):
        
        if m == NWA.loc[i,'mh2']:
            chunk.loc[len(chunk.index)] = NWA.loc[i]
    row = ATLAS.loc[ATLAS['mH'] == m,'xs']
    if not row.empty: 
        comp = ATLAS.loc[ATLAS['mH'] == m,'xs'].values[0].astype(float)
    sorted_chunk = chunk.sort_values('xs')
    sorted_chunk['ks'] = pd.to_numeric(sorted_chunk['ks'], errors='coerce')
    sorted_chunk['xs'] = pd.to_numeric(sorted_chunk['xs'], errors='coerce')      

    ## Interpolate the k = k(xsec) and use this function to get the correct coupling for the bound
    if len(sorted_chunk) > 0:
        ks_interp = np.interp(comp,sorted_chunk['xs'],sorted_chunk['ks'])
        data_rows.append([m, float(ks_interp),float(comp)])
        
    if m == 350:
        chunk1 = chunk.sort_values('ks')
        chunk1['ks'] = pd.to_numeric(chunk1['ks'], errors='coerce')
        chunk1['xs'] = pd.to_numeric(chunk1['xs'], errors='coerce')
        plt.axhline(y=comp, color='r', linestyle='-', label = 'ATLAS bound for m=350GeV')
        plt.axvline(x=ks_interp, color='green',label='interpolated ks value')
        ax = plt.gca()
        ax.plot(chunk1['ks'],chunk1['xs'], marker='.', markersize='5',color='C0')
        ax.set_xlabel('ks', fontsize=18)
        ax.set_ylabel('cross section (pb)',fontsize=18)
        plt.grid()
        ax.set_yscale('log')
        ax.set_xscale('log')
        ax.legend()
        plt.show()
        

for m in CMS[CMS['mH'] <=190]['mH']:
    chunk = pd.DataFrame(columns=['mh2', 'ks', 'xs'])
    for i in range(len(NWA)):
        if m == NWA.loc[i,'mh2']:
            chunk.loc[len(chunk.index)] = NWA.loc[i]  
           
    row = CMS.loc[CMS['mH'] == m,'xs']
    if not row.empty:
        comp = CMS.loc[CMS['mH'] == m,'xs'].values[0].astype(float)
    sorted_chunk = chunk.sort_values('xs')
    sorted_chunk['ks'] = pd.to_numeric(sorted_chunk['ks'], errors='coerce')
    sorted_chunk['xs'] = pd.to_numeric(sorted_chunk['xs'], errors='coerce')      

    ## Interpolate the k = k(xsec) and use this function to get the correct coupling for the bound
    if len(sorted_chunk) > 0:
        ks_interp = np.interp(comp,sorted_chunk['xs'],sorted_chunk['ks'])
        data_rows.append([m, float(ks_interp),float(comp)])
    
    
final_bound = pd.DataFrame(data_rows, columns=['m','ks','Exp-xsec']) # this is the data to plot, that will give me the right bound
final_bound.head() 

# os.chdir('/gpfs/home/argiannakopo')
# plt.savefig("xsecVsku_down.pdf", format="pdf", bbox_inches="tight")
```

```python
## Make sure that the interpolation process has worked correctly before plotting. 
m_to_remove = []
for m in final_bound['m']:
    mass_chunk = NWA[NWA['mh2'] == m]
    min_xs = mass_chunk['xs'].min()
    max_xs = mass_chunk['xs'].max()
    if float(final_bound[final_bound['m']== m ]['Exp-xsec']) > max_xs or float(final_bound[final_bound['m'] == m]['Exp-xsec']) < min_xs:
        m_to_remove.append(m)

final_bound = final_bound[~final_bound['m'].isin(m_to_remove)]
```
