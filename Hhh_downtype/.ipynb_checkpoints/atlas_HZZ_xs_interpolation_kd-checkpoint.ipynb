```python
import os
import pandas as pd 
import numpy as np 
import csv
import matplotlib.pyplot as plt
import matplotlib
```

```python
BrHZZ = pd.read_csv('atlas_BR_H_to_ZZ_kd.csv')
BrHZZ['mH'] = BrHZZ['mH'].astype(float)
BrHZZ.head()
```

```python
#Copy the {mh2,kd,xsec} results from MG5 Events directory to a dataframe labeled simres
kd = []
mh2 = []
xs = []
# Specify the directory you want to search
root_directory = '/gpfs/scratch/argiannakopo/H_to_ZZ_kd_cosbma0p025_xi0_atlas/Events/'

# Use os.walk to iterate through all subdirectories
for root, dirs, files in os.walk(root_directory):
    for subdir in dirs:
        # Print the full path of each subdirectory
        subdirectory_path = os.path.join(root, subdir)
        os.chdir(subdirectory_path)
    # with gzip.open('unweighted_events.lhe.gz', 'rb') as f_in:
    #     with open('unweighted_events.LHE', 'wb') as f_out:
    #         shutil.copyfileobj(f_in, f_out)
        start1 = 0
        start2 = 0
        with open(subdir+'_tag_1_banner.txt', 'r') as csvfile:

            rows = csv.reader(csvfile, skipinitialspace=True, delimiter=' ')
            for row in rows:
                if len(row) > 1:
                    if row[0] == "BLOCK":
                        start1 = 1
                    if start1 == 1:
                        if row[2] == "#" and row[3] == "kd":
                            kd.append(float(row[1]))
                        if row[0] == "35" and row[3] == "mh2":
                            mh2.append(float(row[1]))
                    if row[1] == "Integrated" and row[2] == "weight":
                        xs.append(float(row[5]))

simres = pd.DataFrame(zip(mh2,kd,xs), columns = ['mh2','kd','xs'])
simres.head()
```

```python
simres['mh2'] = simres['mh2'].round(4)
simres['kd'] = simres['kd'].round(4)
BrHZZ['mH'] = BrHZZ['mH'].round(4)
BrHZZ['kd'] = BrHZZ['kd'].round(4)
```

```python
# NWA: Multiply each row of the simulation results dataframe with the appropriate Br(A -> Zh)

NWA = pd.DataFrame(columns=['mh2','kd','ZZ','xs'])
for i in range(len(simres)):
    HZZ = float(BrHZZ.loc[(BrHZZ['mH'] == simres.loc[i,'mh2']) & (BrHZZ['kd'] == simres.loc[i,'kd']),'ZZ'])
    xsec = float(simres.loc[i,'xs'])
    NWA.loc[i,'xs'] = HZZ * xsec
    NWA.loc[i,'mh2'] = simres.loc[i,'mh2']
    NWA.loc[i,'kd'] = simres.loc[i,'kd']
    NWA.loc[i,'ZZ'] = HZZ

NWA.head()
    
```

```python
NWA = NWA[NWA['xs'] != 0 ].reset_index(drop=True)
NWA.head()
```

```python
# Load the experimental bounds to compare with my results 

ATLAS = pd.read_csv('/gpfs/home/argiannakopo/SFV/Exp_data/X-ZZ-ATLAS-2009p14791.csv')
ATLAS.head()
```

```python
### Interpolate based on the values of the narrow width approximation to get final bound

data_rows = []

for m in ATLAS['mH']:
    chunk = pd.DataFrame(columns=['mh2', 'kd', 'xs'])
    for i in range(len(NWA)):
        
        if m == NWA.loc[i,'mh2']:
            chunk.loc[len(chunk.index)] = NWA.loc[i]        
    comp = float(ATLAS.loc[ATLAS['mH'] == m,'xs']) 
    sorted_chunk = chunk.sort_values('xs')
    sorted_chunk['kd'] = pd.to_numeric(sorted_chunk['kd'], errors='coerce')
    sorted_chunk['xs'] = pd.to_numeric(sorted_chunk['xs'], errors='coerce')      

    ## Interpolate the k = k(xsec) and use this function to get the correct coupling for the bound
    if len(sorted_chunk) > 0:
        kd_interp = np.interp(comp,sorted_chunk['xs'],sorted_chunk['kd'])
        data_rows.append([m, float(kd_interp),float(comp)])
        
    if m == 350:
        chunk1 = chunk.sort_values('kd')
        chunk1['kd'] = pd.to_numeric(chunk1['kd'], errors='coerce')
        chunk1['xs'] = pd.to_numeric(chunk1['xs'], errors='coerce')
        plt.axhline(y=comp, color='r', linestyle='-', label = 'ATLAS bound for m=350GeV')
        plt.axvline(x=kd_interp, color='green',label='interpolated kd value')
        ax = plt.gca()
        ax.plot(chunk1['kd'],chunk1['xs'], marker='.', markersize='5',color='C0')
        ax.set_xlabel('kd', fontsize=18)
        ax.set_ylabel('cross section (pb)',fontsize=18)
        plt.grid()
        ax.set_yscale('log')
        ax.set_xscale('log')
        ax.legend()
        plt.show()
        

final_bound = pd.DataFrame(data_rows, columns=['m','kd','ATLASxsec']) # this is the data to plot, that will give me the right bound
final_bound.head() 

# os.chdir('/gpfs/home/argiannakopo')
# plt.savefig("xsecVsku_down.pdf", format="pdf", bbox_inches="tight")
```

```python
## Make sure that the interpolation process has worked correctly before plotting. 
m_to_remove = []
for m in final_bound['m']:
    mass_chunk = NWA[NWA['mh2'] == m]
    min_xs = mass_chunk['xs'].min()
    max_xs = mass_chunk['xs'].max()
    if float(final_bound[final_bound['m']== m ]['ATLASxsec']) > max_xs or float(final_bound[final_bound['m'] == m]['ATLASxsec']) < min_xs:
        m_to_remove.append(m)

final_bound = final_bound[~final_bound['m'].isin(m_to_remove)]
```
